```python
raw_data['has_locations'] = raw_data.locations.apply(lambda x: x != [])

# 筛选包含'地区'字段的用户
data_name_locations = raw_data.loc[raw_data['has_locations'] == True, ('name','locations')]

def calc_citylist(list):
    res = []
    for item in list:
        res.append(item['name'])
    return res

# 提取'地区'名
data_name_locations['city'] = data_name_locations.locations.apply(lambda x: calc_citylist(x))
data_name_locations.reset_index(inplace = True)
data_name_locations.drop(['index', 'locations'], axis = 1, inplace = True)
data_name_locations
```

```python
import json

with open("name_city.json",'r') as load_f:
    load_dict = json.load(load_f)

new_dict = {}
for _, name_city in load_dict.items():
    name, cities = name_city['name'], name_city['city']
    new_cities = set() # set去重
    for city in cities:
        # 删除 省/市 字符
        city = city.replace('省', '')
        city = city.replace('市', '')
        # 市 -> 省
        if city in reverse_city_dict.keys():
            new_cities.add(reverse_city_dict[city])
        else:
            flag = False
            # 如果'地区'名包含 省/市，替换
            for k, v in reverse_city_dict.items():
                if k in city or v in city:
                    new_cities.add(v)
                    flag = True
                    break
            if flag == False:
                new_cities.add(city)

    new_dict[name] = new_cities
```

```python
def set_default(obj):
    if isinstance(obj, set):
        return list(obj)
    raise TypeError

# save JSON file
with open("new_name_city.json","w") as f:
    json.dump(new_dict, f, indent=4, ensure_ascii=False, default=set_default)
```
