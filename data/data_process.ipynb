```python
import pandas as pd

raw_data = pd.read_json("items.json")
raw_data.describe()
```

```python
raw_data.iloc[0]
```

```python
data_name_follower_count_NAME = "data_name_follower_count.json"

data_name_follower_count = raw_data[["name", "follower_count"]]
data_name_follower_count = data_name_follower_count.sort_values('follower_count', ascending=False)
data_name_follower_count = data_name_follower_count.reset_index(drop=True)
with open(data_name_follower_count_NAME, 'w', encoding='utf-8') as file:
  data_name_follower_count.to_json(file, force_ascii=False)
```
